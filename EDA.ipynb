Data Aggregation, filtering, sorting and grouping

```python
#Creating a DF

df  = pd.DataFrame({"Name":["Anna","Karen","John","Alice","Kevin", "Sanna","Bob","Emily"],
                    "Age":[35,30,57,65,25,19,20,65],
                    "Salary":[20000,60000,145000,170000,30000,10000,220000,120000],
                    "Department":["Tech","Tech","Tech","Healthcare","Operations","Operations","Tech","Tech"]})
```

```python
#Ascengind: Smallest values at the top and values increasing
df.sort_values('Age',ascending=False)#self, by, axys, ascenging
```

```python
df.sort_values("Salary", ascending=True)
```

```python
#Grouping
df.groupby('Department').count()
```

```python
df.groupby('Department')["Name"].count()
```

```python
df.groupby('Department')["Salary"].mean()
```

```python
df.groupby('Department')["Salary"].min()
```

```python
#Filtering
df[df["Salary"] > 100000] #First if the name of the DF
```

```python
df[(df["Salary"] < 100000 ) & (df["Salary"] > 10000)]
```

Data Merging


```python
#Creating 2 data frames
data1 = pd.DataFrame({"Key":["A","B","C","D","E","F","G"],
                                    "Value1":[1,2,3,4,5,6,7]})


data2 = pd.DataFrame({"Key":["C","D","E","F","G","H"],
                                    "Value2":[8,9,10,11,12,13]})
```

```python
data1
```

```python
data2
```

```python
merge_inner_join = pd.merge(left=data1, right=data2,on='Key',how='inner')
merge_inner_join
```

```python
left_join = pd.merge(left=data1, right=data2,on='Key',how='left')
left_join
```

```python
right_join = pd.merge(left=data1, right=data2,on='Key',how='right')
right_join
```

```python
merged_left_anti = pd.merge(left=data1, right=data2,on='Key',how='left',indicator=True)
merged_left_anti
```
